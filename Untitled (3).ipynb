**Exploratory Data Analysis and Visualization**

```r
mean_table<-tibble(mean_players)
mean_table

ggplot(players,aes(x = played_hours))+
geom_histogram(bins = 10,fill = "blue")+
labs(x = "Played hours", y = "Count")+
ggtitle("Distribution of played hours")

ggplot(players,aes(x = Age))+
geom_histogram(bins = 10,fill = "green")+
labs(x = "Players' age", y = "Count")+
ggtitle("Distribution of Age")

ggplot(players, aes(x = played_hours, y = Age, color = subscribe)) +
 geom_point() +
 labs(x ="Played hours", y = "Players' age")+
 ggtitle("Players'Age & Played hours")



new_players<-players|>
mutate(subscribe = as_factor(subscribe))


#defining knn model
knn_spec<-nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
set_engine("kknn") |>
set_mode("classification")

#defining recipe
players_recipe <- recipe(subscribe ~ Age+ played_hours, data = new_players) |>
step_scale(all_predictors()) |>
step_center(all_predictors())

#Create workflows and fit models
players_fit <- workflow() |>
add_recipe(players_recipe) |>
add_model(knn_spec) |>
fit(data = players) 

#Plot the graph
prediction_plot<-ggplot(players, aes(x = played_hours, y = Age, color = subscribe)) +
 geom_point() +
 labs(x ="Played_hours", y = "Players' age")+
 ggtitle("How Played Hours and Players' age affect subscription")
 prediction_plot
```

**Methods and Plan**
In this case, Knn is appropriate because it doesn't require complex training process, the above graph has nonlinear distribution, this dataset does not have large samples and there are few prediction variables.
We need to assume that the data is scaled and it does not have large scales(large data would make knn model slow)
Large dataset would limit the speed of predictions; knn model will predict based on the majority class, this would result more error in predictions.
Calculating the accuracy of this model in other training and testing dataset to evaluate the performance, using tune() to select the most appropriate K neighbors and finally compare the result with other model.
To process the data and apply the model, First I need to standardize the data. Second, I plan to set up the knn model where neighbors is set to be tune to find the most suitable value.Third, I define the recipe where subscribe is the dependent variable(be predcited), and Age and played_hours are the independent variables. Next, create a workflow and fit models into it. Finally, plot the scatter plot. Our two axis are played_hours and age which would help us predict whether a player subscribes or not.
